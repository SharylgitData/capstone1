```python
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

def preprocess_text(text):
    # Lowercasing
    text = text.lower()
    # Removing non-alphanumeric characters
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
    # Tokenization
    tokens = word_tokenize(text)
    # Stopword Removal
    filtered_tokens = [token for token in tokens if token not in stopwords.words('english')]
    return " ".join(filtered_tokens)
```

```python
from docx import Document

def read_docx(file_path):
    doc = Document(file_path)
    full_text = []
    for para in doc.paragraphs:
        full_text.append(para.text)
    return '\n'.join(full_text)

documents = read_docx('/content/sample_data/Off-Road Vehicles Act.docx')

print(documents)
```

```python
from transformers import BertModel, BertTokenizer
import faiss
import torch
import numpy as np
#distillbert
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')

def embed_text(text):
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model.to(device)
    inputs = tokenizer(text, return_tensors="pt", max_length=512, padding=True, truncation=True).to(device)

    #inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True, max_length=512)
    outputs = model(**inputs)
    return outputs.pooler_output.detach().numpy()  # Using the pooled output

# Assume `documents` is a list of preprocessed texts
document_embeddings = np.vstack([embed_text(doc) for doc in documents])
faiss_index = faiss.IndexFlatL2(document_embeddings.shape[1])
faiss_index.add(document_embeddings)

def retrieve_documents(query, k=5):
    query_embedding = embed_text(preprocess_text(query))
    distances, indices = faiss_index.search(query_embedding, k)
    return [documents[idx] for idx in indices[0]]
```

```python
from transformers import T5ForConditionalGeneration, T5Tokenizer, GPT2LMHeadModel, GPT2Tokenizer

# Initialize T5
t5_tokenizer = T5Tokenizer.from_pretrained('t5-base')
t5_model = T5ForConditionalGeneration.from_pretrained('t5-base')

def generate_response_with_t5(query):
    inputs = t5_tokenizer.encode("summarize: " + query, return_tensors="pt", max_length=512)
    outputs = t5_model.generate(inputs, max_length=200, num_beams=5, early_stopping=True)
    return t5_tokenizer.decode(outputs[0], skip_special_tokens=True)

# Initialize GPT-4 as GPT-2 for demo purposes (GPT-4 access requires OpenAI API)
gpt2_tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
gpt2_model = GPT2LMHeadModel.from_pretrained('gpt2')

def refine_with_gpt2(text):
    inputs = gpt2_tokenizer.encode(text, return_tensors="pt")
    outputs = gpt2_model.generate(inputs, max_length=300, num_beams=5, early_stopping=True)
    return gpt2_tokenizer.decode(outputs[0], skip_special_tokens=True)
```

```python
def query_system(query):
    retrieved_docs = retrieve_documents(query)
    combined_context = " ".join(retrieved_docs)
    response = generate_response_with_t5(combined_context)
    refined_response = refine_with_gpt2(response)
    return refined_response
```
